## Notation

$\mu$ $\quad$ reduced mass

$T = \frac{1}{f}$, $f$ frequency

$\nu$ $\quad$ frequency (number of oscillations), Greek letter nu

$\lambda$ wavelength

$c = \lambda \nu$, in vacuum

$\omega =2\pi f$, angular frequency

$k= \frac{2\pi}{\lambda}$, spatial analog of $\omega$

Einstein eqquation (Energy of a photon): $E = h \nu = \hbar \omega$

Planck constant  
$h≈6.626\times 10^{-34}$Js

reduced Planck constant or Dirac constant  
$\hbar = \frac{h}{2 \pi}$

## physics

Spin 1/2 particles are representated by orthogonal vectors
in the two dimensional complex Hilbert space.

A magnetic field is the magnetic effect of electric currents
and magnetic materials. The magnetic field at any given
point is specified by both a direction and a magnitude (or
trength); as such it is a vector field.

An electric field is a vector field that associates to each
point in space the Coulomb force experienced by a unit
electric charge. Electric fields converge and diverge at
electric charges and they can be induced by time-varying
magnetic fields. The electric field combines with the
magnetic field to form the electromagnetic field.

An electromagnetic field (also EM field) is a physical field
produced by electrically charged objects. It affects the
behavior of charged objects in the vicinity of the field.
The electromagnetic field extends indefinitely throughout
space and describes the electromagnetic interaction. It is
one of the four fundamental forces of nature (the others are
gravitation, weak interaction and strong interaction).

Maxwell's Equations together with the Lorentz force law
form a complete description of classical electrodynamics (or
electromagnetism) including both electricity and magnetism.

(Electromagnetics) An **isotropic** medium is one such that the permittivity,
ε, and permeability, μ, of the medium are uniform in all directions of the
medium, the simplest instance being free space.

### Particles

fermions:
Particles with half-integer spin.

boson:
particles with integer spin.

$^3\ce{He}$ has spin 1/2 and is therefore a fermion.
$^4\ce{He}$ has spin 0 and is a boson.

Half-integer spin means that the intrinsic
angular momentum value of fermions is $\hbar$
times a half integer (1/2, 3/2, 5/2, etc.).

### Semiconductors

DOS: density of states.

The frontier electronic structure of solids is characterized by valence
and conduction bands, which are analogous to the HOMO and LUMO
of molecules built from MO theory.

$\ce{Si}$ can be produced very pure.

The p-n junction is an electrical diode,
a device though which current flows in
only one direction.

p-n diodes are the basis for solar cells,
photodiodes (light detectors), light emitting
diodes (LED), and laser diodes.

drift:
Current generated by an electric field.

Net diffusion occurs until balanced by
an electric field at the junction.
At this point, equilibrium is
established (Fermi level equal).
