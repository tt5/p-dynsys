$Ax=\lambda x$ $\quad (A-\lambda I)x=0$

$\det (A-\lambda I)=0$

eigenvalues
$\lambda_i=\lambda_1,...,\lambda_n$

$\operatorname{trace}(A)=\sum_{i=1}^{n}\lambda_i$

$\det(A)=\prod_{i=1}^{n}\lambda_i$

$A^k$ has $\lambda_i^k$

the eigenvalues of $A^{-1}$ are $\frac{1}{\lambda}$
