## Euler's Totient Function

$$\varphi(n) := |\{ a\in\mathbb{N} : 1\leq a \leq n \wedge \text{gcd}(a,n)=1 \}|$$

$\varphi(n^m) = n^{m-1} \varphi(n)$

$\varphi(p) = p - 1$

$\varphi(36) = \varphi(2^2 3^2) = 36(1-\frac{1}{2})(1-\frac{1}{3}) = 36\cdot \frac{1}{2} \cdot \frac{2}{3} = 12$

half of the 36 integers from 1 to 36 are divisible by 2, leaving 18, a third of those are divisible by 3, leaving 12 that are coprime to 36

If $\text{gcd}(g,n) = 1$ and $g$ is of multiplicative order $\varphi(n)$ modulo $n$,
then $g$ is a **primitive root of $n$**.

If $n$ has a primitive root, then it has exactly $\varphi(\varphi(n))$ of them.

The multiplicative order of $10 \,(\text{mod } 7)$ is $6$, since $10^6 \equiv 1 (\mod 7)$.

A modulo multiplication group is a finite group $M_m$ of residue classes prime to $m$,
under multiplication mod $m$. $M_m$ is Abelian of group order $\varphi(n)$.

If $n$ is composite, there exists a subgroup of the multiplicative group, called
the "group of false witnesses", in which the elements, when raised to the power of $n-1$, are
congruent to 1 mod $n$.

561 is a Carmichael number, thus $s^{560}$ is congruent to 1 mod 561 for any integer $s$ coprime to 561.
The subgroup of false witnesses is, in this case, not proper, it is the entire group of multiplicative units modulo 561, which consists of 320 residues.
