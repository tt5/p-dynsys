# Hopf Bifurcation

### Normal Form

https://en.wikipedia.org/wiki/Hopf_bifurcation

A term is said to be of normal form if it does not admit any further rewrites.


>Normal Form: $$\dot{z}=(\lambda +\mathrm{i}\omega)z+(\alpha + \mathrm{i}\beta)|z|^{2}z$$

The number $Î±$ is called the first **Lyapunov coefficient**.
$\omega$ und $\beta$ are optional.

$$r(t)=r_{0}{\sqrt {\frac {\lambda }{\left(\lambda +\alpha r_{0}^{2}\right)e^{-2\lambda t}-\alpha r_{0}^{2}}}}$$

Writing $z=re^{i\theta}$ (polar coordinates) and considering the real and imaginary parts as distinct, one obtains a pair of ordinary differential equations:

---


$\dot r=\alpha r + a r^{3} + O(r^5)$

$\dot\theta=\omega+ b r^{2} + O(r^4)$

$\alpha = d \mu + O(\mu^2)$


d > 0, 

---

$\dot\theta=1+\beta r^{2} \quad$

$\dot r=\lambda r +\alpha r^{3}$

If $\alpha$ is negative then there is a stable $\color{green}\text{limit cycle}$ for $\lambda >0$.
$r={\sqrt {-\lambda /\alpha }}$ and $\theta =1+\beta r^{2}$.This is the **supercritical** regime [Verlauf].

If $\alpha$ is positive then there is an unstable limit cycle for $\lambda <0$. The *bifurcation* is said to be **subcritical**. 

The appearance or the disappearance of a periodic orbit through a local change in the stability properties of a fixed point is known as the **Hopf bifurcation**.
