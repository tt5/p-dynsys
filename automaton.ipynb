```python
t = np.zeros((60,60), dtype='int64')
s = [1] + [1 for x in range(t.shape[1]-2,0,-1)] + [1]
t[0,] = s
#s = [1] + [math.floor((x+1)*math.sqrt(2)) - math.floor(x*math.sqrt(2)) for x in range(t.shape[1]-2,0,-1)] + [1]
s = [1] + [1 for x in range(t.shape[1]-2,0,-1)] + [1]
t[1,] = s

def processLine(t, l):
    for i in range(6,t.shape[1]-5):
        if t[l-1,i] != 0:
            t[l+1,i] = (math.floor((t[l,i]**2 - t[l,i-1] * t[l,i+1] + 3) // (t[l-1,i])))%5
        else:
            t[l+1,i] = 0

            
for i in range(1,t.shape[0]-1):
    processLine(t,i)

t = t%5

plt.imshow(t, interpolation='nearest');plt.axis('off');plt.show()
```
