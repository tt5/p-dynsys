```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
terrain = np.zeros((3,3), dtype='i')
terrain[1,2] = 1
terrain
```

```python
def processLine(terrain, lineNumber):
    for i in range(terrain.shape[1]-1,-1,-1):
        if terrain[lineNumber,i] == 1:
            terrain[lineNumber+1,i] = 1
            
for i in range(0,terrain.shape[0]-1):
    processLine(terrain,i)
    
plt.imshow(terrain, interpolation='nearest');plt.show()
```
