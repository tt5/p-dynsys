the f.s.a.e.:

- $G = NH$, $N \cap H = \{e\}$
- $\forall g \in G \exists n \in  N$ (unique) and $h\in H$ (unique) s.t. $g=nh$
- $\forall g \in G \exists n \in  N$ (unique) and $h\in H$ (unique) s.t. $g=hn$
